 **1. Calcul parallèle:Écrivez un programme qui utilise les threads pour calculer les carrés
des nombres de 0 à 100 et imprimer les résultats.**


```python
import threading

# Calculer le carré d'un nombre
def calcul_carre(nb):
    carre = nb * nb
    print(f"Le carré de {nb} est {carre}\n")

# Créer 101 threads (pour les nbres de 0 à 100)
# La liste vide threads est créée pour stocker les objets de thread.
threads = []

# La boucle for itère à travers les nombres de 0 à 100.
# Pour chq nombre, un nouvel objet de thread est créé en spécifiant la fonction calculer_carre comme cible,
# et en passant le nombre comme un argument.
for nb in range(101):
    # Chq thread est ajouté à la liste threads et démarre son exécution par la méthode start().
    thread = threading.Thread(target=calcul_carre, args=(nb,))
    threads.append(thread)
    thread.start()

# Après avoir démarré tous les threads, une autre boucle for attend que tous les threads se terminent
# en utilisant la méthode join() afin de garantir que le programme principal attend que tous les threads
# aient terminé avant de continuer.
for thread in threads:
    thread.join()

print("Fin.")
```

 **2. Web Scraping parallèle: Écrivez un programme qui utilise les processus pour
télécharger le contenu de plusieurs URL et imprimer les résultats. Les URL à utiliser seront
fournies.**

```python
import multiprocessing
import requests


# La fonction download_and_print prend une URL en entrée, effectue une demande HTTP pour récupérer
#le contenu de cette URL, ensuite imprime le contenu ou affiche un message d'erreur en cas de problème.
def download_and_print(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            print(f"Le contenu de {url} :")
            print(response.text)
        else:
            print(f"{url} non téléchargé. Statut HTTP : {response.status_code}")
    except Exception as e:
        print(f"Une erreur s'est produite lors du téléchargement de {url} : {str(e)}")

# Dans la section if __name__ == "__main__":, une liste urls est définie, contenant les URL a télécharger.
if __name__ == "__main__":

    # exemple avec 2 URLs
    urls = [
        "https://www.example.org",
        "https://learntocodewith.me/learn/why-computer-science/"

    ]

    # Ensuite, la boucle for itère sur chaque URL dans la liste urls.
    # Pour chaque URL, un nouveau processus est créé en spécifiant la fonction download_and_print
    # comme cible, et en passant l'URL comme argument.
    process_list = []
    for url in urls:
        process = multiprocessing.Process(target=download_and_print, args=(url,))
        # Chaque processus est ajouté à la liste process_list, et démarre son exécution par la méthode start().
        process_list.append(process)
        process.start()

    # Après avoir démarré tous les processus, une autre boucle for attend que tous les processus
    # se terminent en utilisant la méthode join().
    # Pour garantir que le programme principal attend que tous les processus sont terminés avant de continuer.
    for process in process_list:
        process.join()

    print("Fin.")

```

 **3. Synchronisation: Écrivez un programme où deux threads essaient d'accéder à une
ressource partagée. Utilisez un verrou pour éviter un problème de concurrence.**

```python
import threading

# Créer un verrou (lock)
lock = threading.Lock()

# Variable shared_resource est déclarée et initialisée à 0.
# Cette variable représente la ressource partagée entre les threads.
shared_resource= 0

# Fonction increment_whith_lock utilise un verrou pour garantir que l'incrémentation de la variable shared_resource est effectuée
# de manière atomique et que deux threads ne peuvent pas interagir avec la ressource partagée simultanément.
# Cette fonction incrémente la variable shared_resource 10000 fois.
def increment_whith_lock():
    global shared_resource
    for _ in range(100):
        # Utilisation du verrou
        with lock:
            shared_resource += 1

# Deux threads (thread1 et thread2) sont créés en spécifiant la fonction increment_whith_lock comme cible.
thread1 = threading.Thread(target=increment_whith_lock)
thread2 = threading.Thread(target=increment_whith_lock)

# Démarrer les threads par la methode start()
thread1.start()
thread2.start()

# Attendre la fin des threads par la methode join()
thread1.join()
thread2.join()

# Afficher la valeur finale de la ressource partagée
print("Valeur de la ressource partagée finalement :", shared_resource)
```

**Partie 3:
 L'objectif de cet exercice est de comparer le temps d'exécution de l'algorithme de tri par fusion (merge sort) lorsqu'il est exécuté de manière séquentielle, parallélisée avec des threads et parallélisée avec des processus**

```python
import threading
import multiprocessing
import time
import numpy as np

# Fonction de fusion pour le tri par fusion
def merge(left, right):
    result = []
    i, j = 0, 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

# Fonction de tri par fusion
def merge_sort(lst):
    if len(lst) <= 1:
        return lst

    middle = len(lst) // 2
    left = merge_sort(lst[:middle])
    right = merge_sort(lst[middle:])

    return merge(left, right)


# Fonction pour trier une liste en utilisant le tri par fusion de manière séquentielle
def tri_sequentiel(lst):
    return merge_sort(lst)

# Fonction pour trier une liste en utilisant le tri par fusion parallélisé avec des threads
def tri_parallel_avec_thread(lst, thread_count):

    #chunk_size est calculé en divisant la longueur de la liste lst par le nombre de threads spécifié (thread_count).
    #Cela détermine la taille de chaque chunk
    chunk_size = len(lst) // thread_count

    #La liste lst est ensuite divisée en thread_count chunks en utilisant une liste (chunks).
    #chaque chunk contient une portion de la liste à trier.
    chunks = [lst[i:i+chunk_size] for i in range(0, len(lst), chunk_size)]

    threads = []

    #Une liste vide sorted_chunks est créée pour stocker les chunks triés.
    sorted_chunks = []

    #Une boucle for est utilisée pour créer et démarrer des threads pour trier chaque chunk.
    #Chaque thread utilise la fonction merge_sort pour trier son chunk et ajoute le résultat trié
    #à la liste sorted_chunks.
    for chunk in chunks:

        #La fonction lambda sert a passer le résultat trié au thread, dans le but de collecter les résultats
        #de chaque thread.
        thread = threading.Thread(target=lambda x, s: s.append(merge_sort(x)), args=(chunk, sorted_chunks))
        threads.append(thread)
        thread.start()
        #Les threads sont ensuite attendus (join)
        thread.join()

    result = []
    #Les chunks triés stockés dans sorted_chunks sont fusionnés pour obtenir la liste triée finale à renvoyer.
    for chunk in sorted_chunks:
        result = merge(result, chunk)

    return result

# Fonction pour trier une liste en utilisant le tri par fusion parallélisé avec des processus
def tri_parallel_avec_processus(arr, process_count):

    # chunk_size est calculé en divisant la longueur de la liste arr par le nombre de processus
    # spécifié (process_count) afin de détermine la taille de chq chunk
    chunk_size = len(arr) // process_count

    # La liste arr est ensuite divisée en process_count chunks en utilisant une liste en compréhension (chunks).
    # Chq chunk contient une portion de la liste à trier.
    chunks = [arr[i:i+chunk_size] for i in range(0, len(arr), chunk_size)]

    processes = []
    # Une liste partagée sorted_chunks est créée en utilisant multiprocessing.Manager().list().
    # Cette liste sert pour stocker les chunks triés par chq processus.
    sorted_chunks = multiprocessing.Manager().list()

    # Une boucle for est utilisée pour créer et démarrer des processus pour trier chaque chunk.
    # Tout processus utilise la fonction merge_sort pour trier son chunk et ajoute le resultat trié à
    # la liste partagée sorted_chunks.
    for chunk in chunks:
        #Comme dans le cas des threads, la fonction lambda sert a passer le résultat trié au processus.
        process = multiprocessing.Process(target=lambda x, s: s.append(merge_sort(x)), args=(chunk, sorted_chunks))
        processes.append(process)
        process.start()

    #Les processus sont ensuite attendus
    for process in processes:
        process.join()

    # Les chunks triés stockés dans sorted_chunks sont finalement fusionnés pour obtenir la liste
    # triée finale à renvoyer.
    result = []
    for chunk in sorted_chunks:
        result = merge(result, chunk)

    return result

if __name__ == "__main__":
    # Tailles de liste à tester
    tailles_listes = [10**4, 10**5, 10**6]

    for taille in tailles_listes:
        # Générer une liste aléatoire
        liste = np.random.randint(0, 1000000, taille).tolist()

        # Tri séquentiel
        debut = time.time()
        tri_sequentiel(liste.copy())
        temps_sequentiel = time.time() - debut

        # Tri parallélisé avec 2 threads
        debut = time.time()
        tri_parallel_avec_thread(liste.copy(), 2)
        temps_deux_threads = time.time() - debut

        # Tri parallélisé avec 2 processus
        debut = time.time()
        tri_parallel_avec_processus(liste.copy(), 2)
        temps_deux_processus = time.time() - debut

        print(f"Taille de la liste = {taille}")
        print(f"Temps d'exécution séquentiel = {temps_sequentiel} secondes")
        print(f"Temps d'exécution avec 2 threads = {temps_deux_threads} secondes")
        print(f"Temps d'exécution avec 2 processus = {temps_deux_processus} secondes")
        print("-" * 100)
```
